```python
import os
directorio = 'f3'  # Reemplaza esto con la ruta de tu directorio
def contar_archivos(directorio):
    return len([nombre for nombre in os.listdir(directorio) if os.path.isfile(os.path.join(directorio, nombre))])

print('Cantidad total de archivos:', contar_archivos(directorio))

```

```python
import os
import shutil

def mover_archivos(directorio_origen, directorio_destino, cantidad):
    archivos = os.listdir(directorio_origen)
    for i, archivo in enumerate(archivos):
        if i >= cantidad:
            break
        origen = os.path.join(directorio_origen, archivo)
        destino = os.path.join(directorio_destino, archivo)
        shutil.move(origen, destino)

directorio_origen = 'f3'  # Reemplaza esto con la ruta de tu directorio origen
directorio_destino = 'f2'  # Reemplaza esto con la ruta de tu directorio destino
cantidad = 12000  # Cantidad de archivos a mover

mover_archivos(directorio_origen, directorio_destino, cantidad)
```
